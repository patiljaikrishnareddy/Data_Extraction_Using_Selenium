# 6/1. Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

```python
# While executing below code please watch the scrolling part as the loading new content after scrolling may not work

#Extracting the first 500 followers displayed for foodtalkindia
driver.get('https://www.instagram.com/foodtalkindia')
wait=WebDriverWait(driver,10)
followers_btn=wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'Y8-fY')))
followers_btn[1].click()
pop_up_window=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))

followers_foodtalkindia=[]
while len(followers_foodtalkindia)<=500:
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight;', pop_up_window) #scrolling
    time.sleep(3)
    followers_foodtalkindia=driver.find_elements(By.XPATH,'//div[@class="isgrP"]/ul/div/li')
    
followers_foodtalkindia=followers_foodtalkindia[:500]

for i in range(len(followers_foodtalkindia)):
    followers_foodtalkindia[i]= followers_foodtalkindia[i].text.replace('\n',' ').split()[0]

followers_foodtalkindia
```

# 6/2. Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
# While executing below code please watch the scrolling part as the loading new content after scrolling may not work

#Extracting my following

myusername='SAMPLE USERNAME'
driver.get('https://www.instagram.com/'+str(myusername))
wait=WebDriverWait(driver,10)
following_btn=wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'Y8-fY')))
following_btn[2].click()
pop_up_window=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))

my_following=[]
curr_len=0
prev_len=-2**31
while prev_len!=curr_len:
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight;', pop_up_window) #scrolling
    time.sleep(3)
    my_following=driver.find_elements(By.XPATH,'//div[@class="isgrP"]/ul/div/li')
    prev_len=curr_len
    curr_len=len(my_following)
    

for i in range(len(my_following)):
    my_following[i]= my_following[i].text.replace('\n',' ').split()[0]

    
# Due to privacy reasons, I have not printed the results
my_following
```

```python
# Set of people whom I follow but they dont follow back

following_set=set(my_following)
followers_set=set(my_followers)

dont_follow_back_set = following_set - followers_set

dont_follow_back_set
```

```python
# List of people who follow foodtalkindia but who do not follow me

driver.get('https://www.instagram.com/foodtalkindia/')

#if 'followed by' button is found
try:
    followed_by_btn=wait.until(EC.presence_of_element_located((By.XPATH,'//section[@class="wW3k-"]//div/a/span')))
    followed_by_btn.click()
    see_all_followers_btn=wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'See All Followers')))
    see_all_followers_btn.click()

    pop_up_window=driver.find_element(By.CLASS_NAME,'isgrP')
    
    #since all the handles that i follow(which show unfollow button) will be displayed on top,
    #below loop will exit when it finds a follow button on the scrolling window

    while True:
        btn_txt=driver.find_elements(By.XPATH,'//div[@class="isgrP"]/ul/div/li//button')
        if btn_txt[len(btn_txt)-1].text=='Follow':   
            break
        driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight;',pop_up_window)
        time.sleep(3)

    #Extracting foodtalkindia followers handles whom I follow
    
    fti_followers_that_i_follow=[]
    accounts= driver.find_elements(By.XPATH,'//div[@class="isgrP"]//span/a')

    j=0
    for i in accounts:
        if btn_txt[j].text=='Follow':
            break
        fti_followers_that_i_follow.append(i.get_attribute('title'))
        j+=1
    
    fti_followers_that_i_follow_set=set(fti_followers_that_i_follow)
    final_result=dont_follow_back_set.intersection(fti_followers_that_i_follow_set)
    for i in final_result:
        print(i)

#if 'followed by' button is not found
except TimeoutException:
    print('Followed by button not found')
    print('There are NO followers of “foodtalkindia” that I am following but those who don’t follow me')
```
